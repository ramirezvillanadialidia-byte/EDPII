La prueba de frecuencias es una de las pruebas estadísticas más utilizadas para verificar la uniformidad de un conjunto de números pseudoaleatorios. Su objetivo es determinar si los números están distribuidos de manera equitativa a lo largo del intervalo [0, 1].

Nuestras hipotesis fueron:

$$H_0: \text{Los números } r_i \text{ siguen una distribución uniforme en el intervalo [0, 1].}$$

$$H_a: \text{Los números } r_i \text{ no siguen una distribución uniforme.}$$


Tenemos que la frecuencia esperada es:
$$f_e = \frac{n}{k}$$

El estadístico:

$$\chi_0^2 = \sum_{i=1}^{k} \frac{(f_{oi} - f_{ei})^2}{f_{ei}}$$


 COmo un extra añadimos el criterio de aceptacion

 \chi_0^2 < \chi_{\alpha, k-1}^2


```python
import pandas as pd
import numpy as np
from scipy import stats

# 1. Datos de la tabla
datos = [
    0.78961, 0.05230, 0.10699, 0.55877, 0.14151, 0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
    0.80548, 0.82654, 0.29453, 0.20852, 0.42989, 0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
    0.89898, 0.57880, 0.67621, 0.05010, 0.00121, 0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
    0.38618, 0.76910, 0.68334, 0.55170, 0.10850, 0.79982, 0.45679, 0.21631, 0.87616, 0.55743,
    0.58962, 0.33216, 0.03185, 0.61168, 0.09264, 0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
    0.29931, 0.30861, 0.83358, 0.51781, 0.03272, 0.57410, 0.26593, 0.85903, 0.43308, 0.35286,
    0.24000, 0.65559, 0.38507, 0.90829, 0.94187, 0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
    0.54325, 0.62400, 0.09133, 0.41678, 0.33954, 0.58244, 0.85853, 0.88752, 0.33729, 0.15506,
    0.23949, 0.53559, 0.33381, 0.49383, 0.75103, 0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
    0.19147, 0.40644, 0.08128, 0.73435, 0.22724, 0.22287, 0.07281, 0.64183, 0.44267, 0.72102
]

# 2. Intervalos de la tabla
n = len(datos)
k = 5
intervalos = ["(0 - 0.2)", "(0.2 - 0.4)", "(0.4 - 0.6)", "(0.6 - 0.8)", "(0.8 - 1.0)"]
f_esperada = n / k

# 3. Calcular Frecuencia Observada (fo)
fo, bins = np.histogram(datos, bins=np.linspace(0, 1, k+1))

# 4. Creamos la tabla
df = pd.DataFrame({
    'Intervalo': intervalos,
    'fo (Frec. Observada)': fo,
    'fe (Frec. Esperada)': [f_esperada] * k
})

# 5. Calcular la última columna: (fo - fe)^2 / fe

df['(fo - fe)^2 / fe'] = ((df['fo (Frec. Observada)'] - df['fe (Frec. Esperada)'])**2) / df['fe (Frec. Esperada)']

# Mostrar tabla
print(df.to_string(index=False))

# 6. Conclusión
chi_cuadrado_total = df['(fo - fe)^2 / fe'].sum()
alfa = 0.05
chi_critico = stats.chi2.ppf(1 - alfa, k - 1)

print("-" * 30)
print(f"Chi-cuadrado calculado (X^2): {chi_cuadrado_total:.4f}")
print(f"Chi-cuadrado crítico (tabla): {chi_critico:.4f}")

# Hipótesis
# H0: Los números siguen una distribución uniforme.
# Ha: Los números NO siguen una distribución uniforme.

if chi_cuadrado_total < chi_critico:
    print("\nResultado: No se rechaza H0. Los números son uniformes.")
else:
    print("\nResultado: Se rechaza H0. Los números no son uniformes.")
```
