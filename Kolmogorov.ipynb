#Prueba de Kolmogorov - Smirnov #
La prueba de Kolmogorov-Smirnov es un procedimiento estadístico diseñado para determinar si una muestra de datos sigue una distribución de probabilidad específica. En el contexto de los números pseudoaleatorios, se utiliza para verificar la uniformidad en el intervalo [0, 1].


Esta prueba es de tipo no paramétrico y se basa en la comparación de la Función de Distribución Acumulada Empírica F_n(x) con la Función de Distribución Acumulada Teórica F(x). Algo que hay que destacar de esta prueba y es muy importante, es que para poder realizarla hay que organizar los datos de menor a mayor.


Para este problema usamos las formulas vistas, como lo son:

Las hipotesis:

$H_0: \text{Los números } r_i \text{ siguen una distribución uniforme } U(0,1).$

$H_a: \text{Los números } r_i \text{ no siguen una distribución uniforme.}$


Para los estadisticos

$D^+ = \max_{1 \le i \le n} \left\{ \frac{i}{n} - r_i \right\}$

$D^- = \max_{1 \le i \le n} \left\{ r_i - \frac{i-1}{n} \right\}$


El estadistico nos queda como:
$D = \max(D^+, D^-)$


Por último tenemos el criterio de decisión:
$D_{\text{crítico}} = \frac{1.36}{\sqrt{n}}$


```python
import pandas as pd
import numpy as np

# Datos de la tarea anterior (Tabla 3.1)
datos = [
    0.78961, 0.05230, 0.10699, 0.55877, 0.14151, 0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
    0.80548, 0.82654, 0.29453, 0.20852, 0.42989, 0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
    0.89898, 0.57880, 0.67621, 0.05010, 0.00121, 0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
    0.38618, 0.76910, 0.68334, 0.55170, 0.10850, 0.79982, 0.45679, 0.21631, 0.87616, 0.55743,
    0.58962, 0.33216, 0.03185, 0.61168, 0.09264, 0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
    0.29931, 0.30861, 0.83358, 0.51781, 0.03272, 0.57410, 0.26593, 0.85903, 0.43308, 0.35286,
    0.24000, 0.65559, 0.38507, 0.90829, 0.94187, 0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
    0.54325, 0.62400, 0.09133, 0.41678, 0.33954, 0.58244, 0.85853, 0.88752, 0.33729, 0.15506,
    0.23949, 0.53559, 0.33381, 0.49383, 0.75103, 0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
    0.19147, 0.40644, 0.08128, 0.73435, 0.22724, 0.22287, 0.07281, 0.64183, 0.44267, 0.72102
]

# Paso 1: Ordenar los datos
datos_ordenados = sorted(datos)
n = len(datos_ordenados)
i = np.arange(1, n + 1)

# Paso 2: Creamos el Dataframe con los datos ya ordenados
df_ks = pd.DataFrame({
    'i': i,
    'Ri (Ordenados)': datos_ordenados,
    'i/n': i / n,
    '(i-1)/n': (i - 1) / n
})

# Paso 3: Calcular las diferencias D+ y D-
df_ks['D+'] = df_ks['i/n'] - df_ks['Ri (Ordenados)']
df_ks['D-'] = df_ks['Ri (Ordenados)'] - df_ks['(i-1)/n']

# Aqui hacemos que nos muestre los 100 resultados
pd.set_option('display.max_rows', None)

# Paso 4: Imprimimos nuestra tabla
print("TABLA COMPLETA DE CÁLCULOS KOLMOGOROV-SMIRNOV:")
print(df_ks.to_string(index=False))

# Paso 5: Resultados finales
d_max = max(df_ks['D+'].max(), df_ks['D-'].max())
alfa = 0.05
# Valor crítico aproximado para n > 35 (1.36 / sqrt(n))
d_critico = 1.36 / np.sqrt(n)

print("\n" + "="*40)
print(f"Estadístico D calculado: {d_max:.4f}")
print(f"Valor crítico (alfa=0.05): {d_critico:.4f}")
print("-" * 40)

# Hipótesis
# H0: Los datos son uniformes
# Ha: Los datos no son uniformes

if d_max < d_critico:
    print("Conclusión: No se rechaza H0. Los números son uniformes.")
else:
    print("Conclusión: Se rechaza H0. Los números NO son uniformes.")
```
