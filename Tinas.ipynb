```python
import numpy as np


a= 190 #peso en kg mínimo
c= 210 #peso en kg más común
b = 230 #peso en kg máximo
n = 5 #tinas
capacidad_max = 1000
iteraciones = 100000

#cálculo de peso
pesos_tinas = np.random.triangular(a, c, b, size=(iteraciones, n))

#suma total de peso
pesos_totales = np.sum(pesos_tinas, axis=1)

# cálculo de probabilidad
excesos = np.sum(pesos_totales > capacidad_max)
probabilidad = excesos / iteraciones

print(f"Probabilidad de que el peso de las tinas exceda 1T: {100*probabilidad:.5f}%")
print(f"Promedio de peso por viaje: {np.mean(pesos_totales):.4f} kg")
```
