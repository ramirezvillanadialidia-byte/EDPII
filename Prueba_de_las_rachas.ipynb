##La prueba de Rachas
La Prueba de Rachas se utiliza para verificar la independencia de una secuencia de números aleatorios. Analizamos si el orden de aparición de los números sigue un patrón o si es verdaderamente aleatorio basándonos en cuántas veces la secuencia cruza su valor medio.

Las formulas que se ocuparon en este código fueron.

El valor esperado de rachas:

$$E(h) = \frac{2n_1n_2}{n_1 + n_2} + 1$$

La varianza del número de rachas

$$V(h) = \frac{2n_1n_2(2n_1n_2 - n_1 - n_2)}{(n_1 + n_2)^2(n_1 + n_2 - 1)}$$

El estadístico de pruebas:

$$Z_0 = \frac{h - E(h)}{\sqrt{V(h)}}$$


```python
import math

# Datos extraídos de tu tabla (100 valores)
datos = [
    0.78961, 0.05230, 0.10699, 0.55877, 0.14151, 0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
    0.80548, 0.82654, 0.29453, 0.20852, 0.42989, 0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
    0.89898, 0.57880, 0.67621, 0.05010, 0.00121, 0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
    0.38618, 0.76910, 0.68334, 0.55170, 0.10850, 0.79982, 0.45679, 0.21631, 0.87616, 0.55743,
    0.58962, 0.33216, 0.03185, 0.61168, 0.09264, 0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
    0.29931, 0.30861, 0.83358, 0.51781, 0.03272, 0.57410, 0.26593, 0.85903, 0.43308, 0.35286,
    0.24000, 0.65559, 0.38507, 0.90829, 0.94187, 0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
    0.54325, 0.62400, 0.09133, 0.41678, 0.33954, 0.58244, 0.85853, 0.88752, 0.33729, 0.15506,
    0.23949, 0.53559, 0.33381, 0.49383, 0.75103, 0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
    0.19147, 0.40644, 0.08128, 0.73435, 0.22724, 0.22287, 0.07281, 0.64183, 0.44267, 0.72102
]

# Clasificación y conteo
n1 = sum(1 for x in datos if x >= 0.5)
n2 = sum(1 for x in datos if x < 0.5)
n = len(datos)

# Cálculo de rachas (h)
rachas = 1
for i in range(1, n):
    # Si un número es >= 0.5 y el anterior era < 0.5 (o viceversa), hay cambio de racha
    actual_pos = datos[i] >= 0.5
    anterior_pos = datos[i-1] >= 0.5
    if actual_pos != anterior_pos:
        rachas += 1

# Cálculos estadísticos
E_h = ((2 * n1 * n2) / n) + 1
V_h = (2 * n1 * n2 * (2 * n1 * n2 - n)) / (n**2 * (n - 1))
Z_0 = (rachas - E_h) / math.sqrt(V_h)

print(f"Resultados:")
print(f"n1 (positivos): {n1}")
print(f"n2 (negativos): {n2}")
print(f"Número de rachas (h): {rachas}")
print(f"Valor esperado E(h): {E_h:.4f}")
print(f"Z calculado: {Z_0:.4f}")

# Conclusión simple (Estilo estudiante)
if abs(Z_0) < 1.96:
    print("Conclusión: Se acepta que los números son aleatorios.")
else:
    print("Conclusión: Los números no son aleatorios.")
```
